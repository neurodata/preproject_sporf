```python
n = 1500
ps, p = 3, 20
X, y = sp(n, ps, p)
Xtest, ytest = sp(10000, ps, p)
```

```python
A = sporf_ternaryColumns(p,p**2,1)
```

```python
Xt = X @ A
```

```python
plt.imshow(X[0:50,:])
```

```python
plt.imshow(Xt)
```

```python
clf = RFC(n_estimators = 1000, max_depth = None, max_features = 10)
clf.fit(X,y)
yhat = clf.predict(Xtest)
```

```python
error_rate = np.mean(yhat != ytest)
print(error_rate)
```

```python
clfPre = RFC(n_estimators = 1000, max_depth = None, max_features = 10)
clfPre.fit(Xt, y)

yhatPre = clfPre.predict(Xtest @ A)
```

```python
error_ratePre = np.mean(yhatPre != ytest)
print(error_ratePre)
```
